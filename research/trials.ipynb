### downlaod data from google drive

```python
import gdown
```

```python
url = "https://drive.google.com/file/d/1q_Yn9A7uhPRz7ew9d51ucwgBiDIpFgR6/view?usp=drive_link"
```

```python
file_id=url.split("/")[-2]
```

```python
prefix = 'https://drive.google.com/uc?/export=download&id='
gdown.download(prefix+file_id, "kidneyscan_data.zip")
```
